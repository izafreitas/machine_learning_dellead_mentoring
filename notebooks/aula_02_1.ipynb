```python
# Separando a coluna value em média, mínimo e máximo
def valores(value):
  media, minimo, maximo = [], [], []
  for i in range(len(value)):
      media.append(value[i].split('-')[0].split('[')[0])
      minimo.append(value[i].split('-')[0].split('[')[1])
      maximo.append(value[i].split('-')[1]).split('[')[0]
  return media, minino, maximo
data.head(2)
```
