### Árvores de Decisão

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
import graphviz
from sklearn.tree import export_graphviz
```

```python
#header=None
data = pd.read_csv('https://raw.githubusercontent.com/hinessacaminha/mentoring-ml/main/carros-arvores.csv')
data.head()
```

```python
data['buying'] = data['buying'].astype('category')
data['maint'] = data['maint'].astype('category')
data['lug_boot'] = data['lug_boot'].astype('category')
data['safety'] = data['safety'].astype('category')
data['car_accep'] = data['car_accep'].astype('category')
data.dtypes
```

```python
data['maint'] = data['maint'].cat.codes
data['buying'] = data['buying'].cat.codes
data['lug_boot'] = data['lug_boot'].cat.codes
data['safety'] = data['safety'].cat.codes
data['car_accep'] = data['car_accep'].cat.codes
```

```python
data['doors'].replace('5more', 5, inplace=True)
data['persons'].replace('more', 5, inplace=True)
```

```python
data['car_accep'].value_counts()
```

```python
data.head()
```

```python
data['doors'] = data['doors'].astype('int')
data['persons'] = data['persons'].astype('int')
```

```python
data['car_accep'].value_counts()
```

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2)
```

Obs: 1 *Para evitar o sobreajuste nos dados de treinamento, você precisa restringir a liberdade
da Árvore de Decisão durante este treinamento. Como você sabe, isso é chamado de regularização.
Os hiperparâmetros de regularização dependem do algoritmo utilizado,
mas geralmente você pode, pelo menos, restringir a profundidade máxima da Árvore
de Decisão. No Scikit-Learn, isto é controlado pelo hiperparâmetro max_depth (o valor
padrão é None, que significa ilimitado).* **Reduzir `max_depth` regularizará o modelo e
reduzirá, assim, o risco de sobreajuste.** 

Obs: 2 *Aumentar os hiperparâmetros `min_*` ou reduzir os hiperparâmetros `max_*` regularizará o modelo.*

```python
# Restringi a profundidade da árvore
tree = DecisionTreeClassifier(criterion='gini', max_depth=5,random_state=0)
model = tree.fit(x_train, y_train)
predicted = model.predict(x_test)
```

```python
print("Acurácia:", model.score(x_test, y_test))
print("Kappa:", metrics.cohen_kappa_score(y_test, predicted))
```

**Análises** 

Valor da `acurácia` sem o hipermarâmetro `max_depth` = 0.9653179190751445

Valor da `acurácia` com o hipermarâmetro `max_depth` = 0.7658959537572254

Valor da `acurácia` com o hipermarâmetro `criterion='entropy'` permaneceu o mesmo


Valor do `Kappa` sem o hipermarâmetro `max_depth` = 0.922228257815573

Valor do `Kappa` com o hiperparâmetro `max_depth` = 0.4231671674968098

Valor do `Kappa` com o hiperparâmetro `criterion='entropy'` = 0.39084507042253525



*O objetivo é reduzir o risco de sobreajuste*

```python
matrix = metrics.confusion_matrix(y_test, predicted)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(model.score(x_test, y_test))
plt.title(all_sample_title, size = 15);
```

**Análise:**

*Com o hiperparâmêtro para `max_depth=2,` os valores de acerto da matriz de confusão foram* - **de 66, 13, 243, 12, para 39, 0, 226, 0 respectivamente**

*Com o hiperparâmêtro `min_samples_split=2` os valores de acerto da matriz de confusão foram* - **35, 0, 228, 0**

*Com o hiperparâmêtro `criterion='entropy'`,  os valores de acerto da matriz de confusão foram* - **35, 0, 228, 0**

```python
# Return the number of leaves of the decision tree.
model.get_n_leaves()
```

**Análise**

*Ao definir a profundidade da árvore como o hiperparêmetro `max_depth` houve uma grande diferença no retorno de folhas da árvore de decisão, isto é, caiu de* **82 para 3 (após o max_depth=2)**, *pois sem esse parâmetro ou setado como None a expansão vai de forma recursiva até atingir chegar a um nó folha puro*

## Usando todos os parâmetros default da classe DecisionTreeClassifier()

```python
tree = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=None, 
                              min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, 
                              max_features=None, random_state=None, max_leaf_nodes=None, 
                              min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
model = tree.fit(x_train, y_train)
predicted = model.predict(x_test)
```
